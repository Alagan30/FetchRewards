```python
'''
1. Certain columns are existing as dictionaries instead of just values. So we have to convert them into corresponding values.

2. Date columns are in integer format rather than datetime format. So datatype conversions are required.

'''
```

```python
import datetime
time = df['createdDate'][0]['$date']

# Converting one of the date value
dt = datetime.datetime.fromtimestamp(time/1000)
print(dt)
print(type(dt))
```

```python
ID, CD = [], []
for index, row in df.iterrows():
    ID.append(row['_id']['$oid'])
    CD.append(row['createdDate']['$date'])
```

```python
# Checking for duplicates in id and created date
temp = pd.DataFrame(data=zip(ID, CD),columns=['ID','Created_Date'])
temp.groupby(['ID','Created_Date']).size().reset_index().rename(columns={0:'count'})
```
